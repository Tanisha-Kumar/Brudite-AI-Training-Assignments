```python
#Assignment 12
def fact(num):
    fact = 1
    if num < 0:
        print("Invalid number")
        return
    if num == 0 or num == 1:
        return fact
    fact = 1
    while(num>1):
        fact *= num
        num -= 1
    return fact

num = int(input("Enter a number: "))
print("Factorial of", num, " : ", fact(num))
```

```python
#Assignment 14
num = int(input("Entre a number : "))
if num <= 1:
    print("Not Prime")
else:
    for i in range(2, num//2 + 1):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```
