```python
#assignment 9
def transpose(mat:list)-> list:
    tr_mat = []
    for col in range(len(mat[0])):
        tr_row = []
        for row in range(len(mat)):
            tr_row.append(mat[row][col])
        tr_mat.append(tr_row)
    return tr_mat

mat = [[1,6,3,5], [4,6,8,2], [11,4,6,7]]
print("original: ", mat)
print("transpose: ", transpose(mat))
```

```python
#assignment 10
def flatten(L: list)-> list:
    flat_L = []
    for i in range(len(L)):
        for j in range(len(L[i])):
            flat_L.append(L[i][j])  
    return flat_L

print("Nested list:", mat)
print("Flat List:", flatten(mat))
```
