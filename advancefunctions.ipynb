```python
#assignment 1
def fibonacci(n, memo=None):
    if memo is None:
        memo = {}

    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)
    return memo[n]

print(fibonacci(5))
print(fibonacci(1))
print(fibonacci(0))
print(fibonacci(10))
```

```python

#assignment 4
def func(callback, L):
    result = []
    for num in L:
        result.append(callback(num))
    return result

print(func(lambda x: x + 10, [1, 2, 3]))

def square(a):
    return a**2

print(func(square, [4,5,6]))
```

```python
from functools import partial

def multiply(a,b):
    return a*b

mul_2 = partial(multiply, 2)

print(mul_2(2))
print(mul_2(3))
print(mul_2(4))
print(mul_2(5))
```

```python
#assignment 10
def avg(L):
    try:
        sum = 0
        for elem in L:
            sum += elem
        return sum/len(L)
    except Exception as e:
        print(f'Error: {e}')
        return None
    
print(avg([1,2,3,4]))
print(avg([]))
```
